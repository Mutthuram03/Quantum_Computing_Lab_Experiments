```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

qc = QuantumCircuit(1)
qc.x(0)

state = Statevector.from_instruction(qc)

print("X Gate Output:")
print(state)
```

```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(1, 1)
qc.h(0)
qc.measure(0, 0)

print("Hadamard Circuit:")
print(qc.draw())
```

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

qc = QuantumCircuit(1)
qc.x(0)

print("X Gate Circuit:")
print(qc.draw())

state = Statevector.from_instruction(qc)
print("Output:", state.data)
```
