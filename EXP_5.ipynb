```python
from qiskit import QuantumCircuit
from qiskit_aer import Aer

qc = QuantumCircuit(3, 2)

# Inputs A=1, B=1
qc.x(0)
qc.x(1)

# Sum (XOR)
qc.cx(0,1)

# Carry (AND)
qc.ccx(0,1,2)

print("Half Adder Circuit:")
print(qc.draw())

# Measure
qc.measure(1,0)   # Sum
qc.measure(2,1)   # Carry

backend = Aer.get_backend('qasm_simulator')
result = backend.run(qc, shots=1024).result()

print("Output:", result.get_counts())
```

```python
from qiskit import QuantumCircuit
from qiskit_aer import Aer

qc = QuantumCircuit(3, 2)

# Inputs A=1, B=0 (you can change)
qc.x(0)

# Difference (XOR)
qc.cx(0,1)

# Borrow
qc.x(0)
qc.ccx(0,1,2)
qc.x(0)

print("Half Subtractor Circuit:")
print(qc.draw())

# Measure
qc.measure(1,0)   # Difference
qc.measure(2,1)   # Borrow

backend = Aer.get_backend('qasm_simulator')
result = backend.run(qc, shots=1024).result()

print("Output:", result.get_counts())
```
