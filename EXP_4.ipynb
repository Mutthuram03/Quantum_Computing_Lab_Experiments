```python
from qiskit import QuantumCircuit
from qiskit_aer import Aer

qc = QuantumCircuit(3, 3)

# Input: |110>
qc.x(0)
qc.x(1)

# CCNOT gate
qc.ccx(0, 1, 2)

print("CCNOT Circuit:")
print(qc.draw())

# Measurement
qc.measure([0,1,2], [0,1,2])

backend = Aer.get_backend('qasm_simulator')
result = backend.run(qc, shots=1024).result()

print("Output:", result.get_counts())
```

```python
from qiskit import QuantumCircuit
from qiskit_aer import Aer

qc = QuantumCircuit(3, 3)

# Input: control=1, swap qubits
qc.x(0)   # control
qc.x(1)   # data

# CSWAP gate
qc.cswap(0, 1, 2)

print("CSWAP Circuit:")
print(qc.draw())

# Measurement
qc.measure([0,1,2], [0,1,2])

backend = Aer.get_backend('qasm_simulator')
result = backend.run(qc, shots=1024).result()

print("Output:", result.get_counts())
```
